```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from pprint import pprint
import uszipcode
import numpy as numpy
import time
import requests
import json
import warnings
warnings.filterwarnings('ignore')
```

```python
#File to load
cldm_to_load = "data\\Chronic_liver_disease_mortality.csv"
```

```python
# Read csv file
cldm_data = pd.read_csv(cldm_to_load, header=0, thousands=r',', dtype={'DataValueAlt': numpy.float64})
```

```python
cldm_data.groupby(['LocationDesc','YearStart'])['DataValueAlt'].mean()
```

```python
cldm_df = cldm_data.groupby(['LocationDesc','YearStart'])['DataValueAlt'].mean()

avg_mortality = cldm_df.unstack(1)
avg_mortality
```

```python
cldm_df1 =cldm_data.groupby(['YearStart'])['DataValueAlt'].mean()
cldm_df1
```

```python
cldm_df1.plot(kind='pie', y = 'YearStart', autopct='%1.1f%%', 
 startangle=90, shadow=False,  legend = False, fontsize=14)
```

```python
cldm_df1.plot(kind='line', y = 'YearStart',  
  legend = False, fontsize=14)
```
